```python
import random
```

```python
class QuickSelect:
    def __init__(self, arr):
        self.arr = arr
        self.arr_len = len(arr)
        self.comparisons = 0
    # Partition function: Rearranges elements 
    # around a pivot (last element)

    def partition(self, arr, l, r):  
        x = arr[r][1]  
        i = l     

        # Iterate through the subarray
        for j in range(l, r):  
            
            # Move elements <= pivot to the left partition
            if arr[j][1] <= x:  
                arr[i], arr[j] = arr[j], arr[i]  
                i += 1  
            
        # Place the pivot in its correct position
        arr[i], arr[r] = arr[r], arr[i]
        self.comparisons += 1  
        return i  

    # Randomizes the pivot to avoid worst-case performance
    def randomPartition(self, arr, l, r):  
        n = r - l + 1  
        pivot = random.randint(0, n - 1)  
        arr[l + pivot], arr[r] = arr[r], arr[l + pivot]  
        return self.partition(arr, l, r)  

    # function to find the k'th smallest element using QuickSelect
    def quickSelect(self, arr, l, r, k):  
        # Check if k is within the valid range of the current subarray
        if 0 < k <= r - l + 1:  
            
            # Partition the array and get the pivot's final position
            pos = self.randomPartition(arr, l, r)  

            # If pivot is the k'th element, return it
            if pos - l == k - 1:  
                return arr[pos]  

            # If pivot's position is larger than k, search left subarray
            if pos - l > k - 1:  
                return self.quickSelect(arr, l, pos - 1, k)  

            # Otherwise, search right subarray and adjust k 
            # (k is reduced by the size of the left partition)
            return self.quickSelect(arr, pos + 1, r, k - (pos - l + 1))  
        
        # Return infinity for invalid k (error handling)
        return float('inf')  

    def kthSmallest(self, arr, k):
        n = len(arr)
        return self.quickSelect(arr, 0, n - 1, k)

    def USEQuickSelect(self, k):
        return self.kthSmallest(self.arr, k)

    def get_num_of_comparisons(self):
        return self.comparisons

    def tener_lista(self):
        return self.arr
```

```python
class MergeSort:
    def __init__(self, arr):
        self.arr = arr
        self.arr_len = len(arr)
        self.comparisons = 0

    def merge(self, arr, left, mid, right):
        self.comparisons += 1

        n1 = mid - left + 1
        n2 = right - mid

        # Create temp arrays
        L = [0] * n1
        R = [0] * n2

        # Copy data to temp arrays L[] and R[]
        for i in range(n1):
            L[i] = arr[left + i]
        for j in range(n2):
            R[j] = arr[mid + 1 + j]
            
        i = 0  
        j = 0  
        k = left  

        # Merge the temp arrays back
        # into arr[left..right]
        while i < n1 and j < n2:
            if L[i][1] <= R[j][1]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Copy the remaining elements of L[],
        # if there are any
        while i < n1:
            arr[k] = L[i]
            i += 1
            k += 1

        # Copy the remaining elements of R[], 
        # if there are any
        while j < n2:
            arr[k] = R[j]
            j += 1
            k += 1

    def USEMergeSort(self):
        self.mergeSort(self.arr, 0, len(self.arr) - 1)

    def mergeSort(self, arr, left, right):
        if left < right:
            mid = (left + right) // 2

            self.mergeSort(arr, left, mid)
            self.mergeSort(arr, mid + 1, right)
            self.merge(arr, left, mid, right)
    
    def get_num_of_comparisons(self):
        return self.comparisons
```

## Pruebas de rendimiento

```python
import time
import tracemalloc
```

```python
def evaluar_rendimiento(acm, funcion, *args, **kwargs): #acm es booleano para recoger el acumulado
    # 1. Iniciar el rastreador de memoria
    tracemalloc.start()
    
    # 2. Iniciar el cronómetro
    inicio_tiempo = time.perf_counter()
    
    # Ejecutar la función
    resultado = funcion(*args, **kwargs)
    
    # 3. Detener el cronómetro
    fin_tiempo = time.perf_counter()
    
    # 4. Obtener estadísticas de memoria (pico de memoria asignada)
    actual, pico_memoria = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    
    tiempo_transcurrido = fin_tiempo - inicio_tiempo
    pico_memoria = pico_memoria / 1024 / 1024

    if not acm:
        print(f"--- Resultados de Rendimiento ---")
        print(f"Tiempo de ejecución: {tiempo_transcurrido:.6f}$ segundos")
        print(f"Memoria utilizada (pico): {pico_memoria:.2f}$ MB")
    else:
        return [tiempo_transcurrido, pico_memoria]
    
    return resultado
```

```python
def clonarLista(l_lista):
    finalLista = []
    for l in l_lista: #Clonar para la ejecución independiente de cada algoritmo
        finalLista.append(l)
    return finalLista
```

```python
def exportar_df_a_Lista(P_dataFrame, month, year):
    P_dataFrame = P_dataFrame.query('Anio == @year and Mes <= @month').groupby('Customer_Id_')['Order_Item_Total_'].sum().reset_index()
    P_dataFrame = P_dataFrame.rename(columns={'Order_Item_Total_': 'Monto_Total_Gastado'})

    ID_cliente = P_dataFrame["Customer_Id_"].tolist()
    montos = P_dataFrame["Monto_Total_Gastado"].tolist()
    t_tup_lista = []

    for registro in range(len(montos)):
        t_tup = (ID_cliente[registro], montos[registro])
        t_tup_lista.append(t_tup)

    return t_tup_lista
```

```python
def exportar_lista_a_txt(nombre_archivo, lista_tuplas):
    with open(nombre_archivo, 'w', encoding='utf-8') as archivo:
        for tupla in lista_tuplas:
            linea = ",".join(map(str, tupla))
            archivo.write(linea + "\n")
```

```python
def graficar(ejeX, Q_quick, M_merge, no_Merge):
    plt.rcParams["figure.figsize"] = (20 , 14)
    
    if not no_Merge:
        plt.axis([0, len(Q_quick), 0, len(M_merge)]) 
        plt.ylim(0, 100+max(max(Q_quick), max(M_merge)))
    else: plt.ylim(0, 30)
    
    plt.xlim(0, max(ejeX))

    plt.plot(ejeX, Q_quick, label='quickselect')

    if not no_Merge:
        plt.plot(ejeX, M_merge, label='mergesort')
        y2 = (max(M_merge)*5001)/5236
        x2 = (max(ejeX)*600)/5236
        y1 = (max(M_merge)*5120)/5236
        x1 = (max(ejeX)*600)/5236

        plt.text(x1 , y1, f"MAX: {max(Q_quick)}", fontsize=12, color='blue')
        plt.text(x2 , y2, f"MAX: {max(M_merge)}", fontsize=12, color='orange')

    plt.legend()
```
